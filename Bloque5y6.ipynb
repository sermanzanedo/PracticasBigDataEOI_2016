Ejercicio 2. Carga en un DataFrame este dataset: https://archive.ics.uci.edu/ml/datasets/Census+Income
Estudia si hay variables como el nivel educativo o el país que influyen en que se tuviese un salario mayor o menor de 50K. Puedes simplemente obtener la cuenta de las veces que sucede una u otra cosa de los diferentes grupos.

```python
data2 = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
#print(data2)
columnas = ["Edad","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","mayor/menor 50K"]
len(columnas)
data2.columns = columnas
data2.head()
```

```python
print(data2["mayor/menor 50K"].unique())
data2.groupby("education").count()
menos50 = data2[data2["mayor/menor 50K"]==" <=50K"].groupby("education").count()
print ("sueldo <= 50K", menos50.Edad)
#type(menos50)
mas50= data2[data2["mayor/menor 50K"]==" >50K"].groupby("education").count()
print ("Sueldo mayor 50k", mas50.Edad)
print(len(menos50))
print (len(mas50))
#plt.subplot(224)
menos50.plot( legend=False,figsize=(8,8))
#plt.subplot(221)
mas50.plot(legend=False, figsize=(8,8))
```

```python
data2.head()
#len(data2)
menos50 = data2[data2["mayor/menor 50K"]==" <=50K"].groupby("native-country").count()
#print(menos50.Edad)
mas50=data2[data2["mayor/menor 50K"]==" >50K"].groupby("native-country").count()
print(menos50.Edad,mas50.Edad)
menos50.plot(kind="barh",title="Menos de 50K", legend=False,figsize=(8,8))
mas50.plot(kind="barh", title="mas de 50K",legend=False,figsize=(8,8))
```
